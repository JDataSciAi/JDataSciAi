```python
from fastapi import FastAPI, UploadFile
import whisper
from gtts import gTTS
import uvicorn

app = FastAPI()
model = whisper.load_model("base")

# Rule-based intent classifier
def classify_intent(text):
    text = text.lower()
    if "refund" in text:
        return "refund"
    elif "cancel" in text:
        return "cancellation"
    elif "order" in text:
        return "order_status"
    else:
        return "unknown"

# Response mapping
responses = {
    "refund": "I understand you’d like a refund. Please provide your order ID.",
    "cancellation": "Your subscription can be cancelled. Would you like me to proceed?",
    "order_status": "Please share your order number to check the status.",
    "unknown": "Sorry, I didn’t understand. Could you rephrase?"
}

@app.post("/voicebot")
async def voicebot(file: UploadFile):
    # Step 1: Save audio
    audio_path = f"temp.wav"
    with open(audio_path, "wb") as f:
        f.write(await file.read())

    # Step 2: Transcribe
    result = model.transcribe(audio_path)
    text = result["text"]

    # Step 3: Classify intent
    intent = classify_intent(text)

    # Step 4: Generate response
    response_text = responses[intent]

    # Step 5: Convert to speech
    tts = gTTS(response_text)
    output_path = "response.mp3"
    tts.save(output_path)

    return {"transcribed_text": text, "intent": intent, "response": response_text, "audio_file": output_path}

if __name__ == "__main__":
    import nest_asyncio
    nest_asyncio.apply()
    import uvicorn
    import asyncio
    from fastapi import FastAPI
    app = FastAPI()
    @app.get("/")
    
    def read_root():
        return {"message": "Hello from FastAPI in Jupyter!"}

config = uvicorn.Config(app, host="0.0.0.0", port=8000, reload=False)
server = uvicorn.Server(config)

loop = asyncio.get_event_loop()
loop.create_task(server.serve())
```
